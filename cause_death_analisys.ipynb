```python
deaths_by_year = df.groupby('Year')[causes].sum()
print(deaths_by_year.head())
```

```python
years = deaths_by_year.index
deaths = deaths_by_year.sum(axis=1)

plt.figure(figsize=(10, 5))
plt.bar(years, deaths, color='red')
plt.xlabel('Year')
plt.ylabel('Number of Deaths')
plt.title('Number of Deaths by Year')
plt.show()
```
